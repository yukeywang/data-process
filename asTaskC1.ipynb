```python
import socket
import csv
import datetime
import time

from time import sleep

host = 'localhost'
port = 9999

s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
s.bind((host, port))
s.listen(1)
print('Listening for a connection...')

csv1 = open('ClimateData-Part2.csv', 'r')
climate = csv.reader(csv1) 
        
csv2 = open('FireData-Part2.csv', 'r')
fire = csv.reader(csv2)

sock, addr = s.accept()
    
print('Accept new connection from...', addr)
cc = next(climate,[])
ff = next(fire, [])
fnumber = 0

while ff != []:                    
    cc = next(climate,[])        

    time = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    if cc != []:
        for i in range(len(cc)):
            cc[i] = cc[i].strip()                
        cc.append(time)
        sendClimate = (",".join(map(str,cc)) + '\n').encode()
        print('Sending climate data: ',sendClimate)
        sock.send(sendClimate)

    i = 0
    while i < 5:
        ff = next(fire, [])
        if ff != []:
            fnumber = fnumber + 1
            mTime =  datetime.datetime.now()
            ff.append(time)
            ff.append(mTime)
            sendFire = (",".join(map(str,ff)) + '\n').encode()
            print('Sending fire data: ', sendFire)
            sock.send(sendFire)
            i = i+1
            print(fnumber)

    sleep(1)
        
sock.close()
print('Connection closed.')
```
